# Data proc

## Proc 1

```python
for index, row0 in enumerate(data2):
    # row0 = data2[0]
    row1 = '\n'.join(row0.split('\n')[15:])

    round_info = row0.split('\n')[10:13]

    songs = [
        i.strip()
        for i in '\n'.join(row0.split('\n')[15:]).split('from Spotify')]
    songs
    songs_names = [songs[0]] + ['\n'.join(i.split('\n')[-3:]) for i in songs[1:-1]]
    songs_names
    songs_votes_voters_rank_person = [
        '\n'.join(i.split('\n')[:4])
        for i in songs[1:]
    ]
    songs_votes_voters_rank_person
    votes_and_comments = [
        '\n'.join(i.split('\n')[4:-3])
        for i in songs[1:]
    ]
    votes_and_comments = [i for i in votes_and_comments[:-1]] + \
        ['\n'.join(votes_and_comments[-1].split('\n')[:-2])]
    votes_and_comments
    votes_and_comments2 = [
        i if i.split('\n')[0] in people else '\n'.join(i.split('\n')[1:])
        for i in votes_and_comments
    ]
    votes_and_comments2
    '|'.join(people)
    person_submit_comments = [
        '' if i.split('\n')[0] in people else i.split('\n')[0]
        for i in votes_and_comments
    ]
    person_submit_comments

    songs
    songs_votes_voters_rank_person
    df_round = pd.DataFrame({
        'Round': round_info[0],
        'Round name': round_info[1],
        'Round description': round_info[2],
        'Song name': songs_names,
        'Song votes':                    [i.split('\n')[0] for i in songs_votes_voters_rank_person],
        'Song voters':                   [i.split('\n')[1] for i in songs_votes_voters_rank_person],
        'Song rank':                     [i.split('\n')[2] for i in songs_votes_voters_rank_person],
        'Song person submitted':         [i.split('\n')[3] for i in songs_votes_voters_rank_person],
        'Song person submitted comment': person_submit_comments,
        'Votes and comments':            votes_and_comments2
    })

    if index == 0:
        df_overall = df_round
    else:
        df_overall = pd.concat(
            [df_overall, df_round], 
            axis = 'rows',
            sort = False
        )
```

## Proc 2

Trim some fields and separate the field `Song name` into `Song`, `Artist`, and `Album`

```python
df_overall_2 = df_overall.copy(deep = True)

df_overall_2['Round'] = df_overall_2['Round'].apply(
    lambda x: x.split(' ')[-1]
)

df_overall_2['Song name'] = df_overall_2['Song name'].str.strip('My song(s)\n')

df_overall_2[['Song', 'Artist', 'Album']] = df_overall_2['Song name'].str.split('\n', expand = True)
df_overall_2 = df_overall_2.drop(columns = ['Song name'])

df_overall_2['Song voters'] = df_overall_2['Song voters'].apply(
    lambda x: x.split(' ')[0]
)

df_overall_2 = df_overall_2[[
    'Round', 'Round name', 'Round description', 
    'Song', 'Artist', 'Album',
    'Song votes', 'Song voters', 'Song rank', 'Song person submitted', 'Song person submitted comment',
    'Votes and comments'
    ]]

df_overall_2
```

```python
df_overall_2.to_csv('output/2_processed_data/260321_proc_data_2.csv', index = False)
```

## Proc 3

Final processing:
- Splitting the hardest column, `Votes and comments`, into row per individual person/comment in the long format.

```python
def proc1(x):
    name_rest = ''

    for index, i in enumerate(x.split('\n')):
        # print(i)
        
        if i in people:
            name_rest += f"<br>{i}"
        else:
            name_rest += f"___{i}"
    
    name_rest = [
        i 
        for i in name_rest.split('<br>')
        if i
    ]
    return name_rest
```

```python
def proc2(x):
    x2 = x.split('___')
    vote = [i for i in x2 if i.isdigit()]
    vote = '' if len(vote) == 0 else vote[0]
    
    comment = '___'.join([i for i in x2 if not i.isdigit()])
    comment = '' if not comment else comment
    return ':::'.join([vote, comment])
```

```python
df_overall_3 = df_overall_2.copy(deep = True)

df_overall_3['Votes and comments'] = df_overall_3['Votes and comments'].apply(
    lambda x: proc1(x)
)
df_overall_3 = df_overall_3.explode('Votes and comments')

df_overall_3['person who voted'] = df_overall_3['Votes and comments'].apply(
    lambda x: x.split('___')[0]
)
df_overall_3['person who voted other'] = df_overall_3['Votes and comments'].apply(
    lambda x: '___'.join(x.split('___')[1:])
)

df_overall_3 = df_overall_3.drop('Votes and comments', axis = 1)

df_overall_3['person who voted other'] = df_overall_3['person who voted other'].apply(
    lambda x: proc2(x)
)
df_overall_3[['person who voted - vote', 'person who voted - comment']] = df_overall_3['person who voted other'].str.split(':::', expand = True)
df_overall_3 = df_overall_3.drop('person who voted other', axis = 1)

df_overall_3
```
